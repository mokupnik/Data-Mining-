```python
def dist(a,b,x,y,c):
    return np.abs(a*x + b*y + c) / np.sqrt (a **2 + b **2)

X = np.arange(1,11)
Y = [10, 10, 11, 12, 18, 18, 19, 26, 19, 26]
plt.plot(X,Y,'b.')
plt.plot(X,2*X+5, 'red')
plt.savefig('zadanie3a.png')
points = list(zip(X,Y))
distance = [dist(2,-1,x,y,5) for x,y in points]
distsum = np.sum(distance)
mina=2
minc=5
print('Stara suma = ', distsum)
    
for a in np.arange(-10,10,0.1):
    for c in np.arange(-10,10,0.1):
        testdistance = [dist(a,-1,x,y,c) for x,y in points]
        newsum = np.sum(testdistance)
        if  newsum < distsum:
            mina=a
            minc=c
            distsum = newsum
print('Nowa suma = ', distsum)           
plt.plot(X,mina*X + minc, 'green')
plt.savefig('zadanie3b.png')
```
